```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load the human detection model
model = YOLO("best.pt")
```

```python
# Perform detection
results = model("Untitled design (6).mp4", conf=0.70, save=True)

# Display the output
for r in results:
    img = r.plot()  # image array with bounding boxes drawn
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
```
